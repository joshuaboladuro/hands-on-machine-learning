Try to build a classifier for the MNIST dataset that achieves over 97% accuracy on the test set.

- Split into train (first 60,000) and test (last 10,000)
- Run `GridSearchCV` over `n_neighbors` in roughly `[3, 4, 5]` and `weights` in `['uniform', 'distance']`
- Measure final accuracy on the held-out test set

## Setup


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)
```

## Solution

_Start writing your solution here._


```python
#Get and inspect the data
from sklearn.datasets import fetch_openml

mnist = fetch_openml('mnist_784', as_frame=False)
x,y = mnist.data, mnist.target

print(x.shape, y.shape)
```

```python
#split data into training and test sets
x_train = x[:60000]
y_train = y[:60000]
x_test = x[60000:]
y_test = y[60000:]

print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```python
#Select and train KNN model 

#Manual training score check only. This is not a final evaluation.
from sklearn.neighbors import KNeighborsClassifier
knn_clf_n3_d = KNeighborsClassifier(n_neighbors=3, weights='distance')
knn_clf_n3_d.fit(x_train, y_train)
print('neighbours: 3, weights: distance, accuracy: ' + str(knn_clf_n3_d.score(x_train, y_train)))

knn_clf_n4_u = KNeighborsClassifier(n_neighbors=4, weights='uniform')
knn_clf_n4_u.fit(x_train, y_train)
print('neighbours: 4, weights: uniform, accuracy: ' + str(knn_clf_n4_u.score(x_train, y_train)))

#Using cross validation
from sklearn.model_selection import GridSearchCV
knn_clf_cv = KNeighborsClassifier()
param_grid = [
    {'n_neighbors': [3,4,5], 
    'weights': ['uniform', 'distance']}
]
grid_search = GridSearchCV(knn_clf_cv, param_grid, cv=5, scoring='accuracy')
grid_search.fit(x_train, y_train)
print('Best parameters: ' + str(grid_search.best_params_))
print('The best model has an accuracy of: ' + str(grid_search.best_score_))
best_model = grid_search.best_estimator_
```

```python
#Test the best model on the test set
best_model.score(x_test, y_test)
print('The best model has a test accuracy of: ' + str(best_model.score(x_test, y_test)))



```
